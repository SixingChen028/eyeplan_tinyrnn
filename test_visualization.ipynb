```python
graph = Graph(num_node = 11, reward_set = np.array([-8, -4, -2, -1, 1, 2, 4, 8]))
graph.reset()

print(graph.child_dict)
print(graph.rewards)
```

```python
visualizer = Visualizer()
visualizer.plot(graph.child_dict, graph.rewards)
```

```python
rollout = []
node = graph.root_node
while not node in graph.leaf_nodes:
    child = np.random.choice(graph.successors(node), size = 1)[0]
    rollout.append((node, child))
    node = child

visualizer.plot(graph.child_dict, graph.rewards, arrows = rollout)
```
